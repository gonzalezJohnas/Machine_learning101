# Titanic Dataset : Example data analysis

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

Load our data into Dataframes

```python
df_train = pd.read_csv('./dataset/titanic/train.csv')
df_test = pd.read_csv('./dataset/titanic/test.csv')

print(df_train.dtypes)
print()
print(df_test.dtypes)
```

# From the dataset description we have 

survival: Survival (0 = No; 1 = Yes)

pclass: Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)

name: Name

sex: Sex

age: Age

sibsp: Number of Siblings/Spouses Aboard

parch: Number of Parents/Children Aboard

ticket: Ticket Number

fare: Passenger Fare

cabin: Cabin

embarked: Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

Lets recapitulate by printing our data and dtypes

We can see which features have nan values

```python
df_train.isna().sum()
```

We can see that the cabin features as 687 missing value over 891 data points so we can remove it

```python
df_train = df_train.drop(['Cabin'] , axis=1)
df_test = df_test.drop(['Cabin'] , axis=1)
print("New shape of our data {}".format(df_train.shape))
df_train.head()
```

We can also get ride off of unnecessary features as Name and PassengerId

```python
df_train = df_train.drop(['Name', 'PassengerId'] , axis=1)
df_test = df_test.drop(['Name', 'PassengerId'] , axis=1)
```

We can have quick statistic from the dataset

```python
df_train.describe()
```

What to do with the NaN values with the Age features ?

```python
df_train = df_train.dropna(subset=['Age', 'Embarked'])
df_test = df_test.dropna(subset=['Age', 'Embarked'])

df_train.shape
df_train.columns
```

Count  the number of survivors

```python
df_train['Survived'].value_counts()
```

Visualize the number of survivors

```python
sns.countplot(df_train['Survived'], label='Count')
```

We can now do a quick epxloratory analysis 

```python
df_train.hist(figsize=(20,10))
```

```python
df_train['Embarked'] = df_train['Embarked'].astype(str)
df_test['Embarked'] = df_test['Embarked'].astype(str)
```

```python
features = ['Age', 'Embarked', 'Sex', 'Pclass', 'SibSp', 'Parch']

fig, axs = plt.subplots(2, 3, figsize=(20, 15))

for r in range(0, 2):
    for c in range(0, 3):
        index = r * 3 + c
        ax = axs[r][c]
        sns.countplot(df_train[features[index]],  hue=df_train['Survived'], ax=ax)
        ax.set_title(features[index])
        ax.legend(title='survived', loc='upper right')
```

We can see that for age the plotting is difficult to intrepet so we can create extra columns and divide people in three categorie : ['child', 'adult', 'senior'] 

```python
def to_age_class(x):
    if x < 18:
        return 'child' 
    elif x > 50:
        return 'senior'
    else:
        return'adult'

df_train['Age_class'] = df_train['Age'].apply(to_age_class)
df_test['Age_class'] = df_test['Age'].apply(to_age_class)


```

```python
features = ['Age_class', 'Embarked', 'Sex', 'Pclass', 'SibSp', 'Parch']

fig, axs = plt.subplots(2, 3, figsize=(20, 15))

for r in range(0, 2):
    for c in range(0, 3):
        index = r * 3 + c
        ax = axs[r][c]
        sns.countplot(df_train[features[index]],  hue=df_train['Survived'], ax=ax)
        ax.set_title(features[index])
        ax.legend(title='survived', loc='upper right')
```

Extra analysis with percentage of survival according to sex and ticket class

```python
df_train.pivot_table('Survived', index='Age_class', columns='Pclass')
sns.barplot(x='Pclass', y='Survived', data=df_train)
```
